## Introduction

The objective of this project is to implement the GOR method for secondary structure prediction and to analyse its quality using two different techniques, namely Q3 and MCC.

### The protein structure
The protein structure represents the **three-dimensional arrangement of atoms** in the amino acids chain. As explained briefly in the first part of this project, it can be described on multiple levels, **4 levels** to be exact: **the primary structure**, **the secondary structure**, **the ternary structure** and lastly, **the quaternary structure**. 

**The primary structure** represents the succession of residus in the polypeptide chain and it is determined by the peptide bonds. This ordered succession codes for all the characteristics of the living organisms. It is unique to one protein and only one and it also can be used to predict, to a large extent, the higher level structures, especially the tertiary structure, but also local features, such as the secondary structure.

**The secondary structure** is composed of highly regular local substructures of the polypeptide backbone chain, namely the **Alpha Helices** and the **Beta Strands/ Sheets** or less frequently **Beta Turns** and **Omega Loops**. These substructures are held together by hydrogen bonds and their geometry is constrained by the specific values of the angles $ \phi $ and $ \psi $. There are multiple ways of experimentally determining the secondary structure of a protein, namely **X-ray crystallography**, **Nuclear magnetic resonance spectroscopy** etc. The previously mentioned X-ray crystallography determines the atomic and molecular structure by measuring the angles and the intensities of the beams diffracted by the analysed crystal. It is the most used method for structure determination and works great for large protein crystal. On the hand, for smaller scale proteins, the NMR (Nuclear magnetic resonance) technique gives better results. As the name suggests, the latter uses nuclear magnetic resonance and exploits the quantum mechanical properties of the atom kernel.

**The tertiary structure** refers to the three dimensional structure of the protein, the folding of Alpha Helices and Beta Strands into a compact globular form. In contrast to the secondary structure, which was driven by hydrogen bonds and $\phi$/$\psi$ angles, the tertiary structure is driven by the hydrophobic properties. In other words, the tertiary structure folds the protein in such a manner that the hydrophobic regions are inside the protein, while the hydrophilic ones are toward the outside zones.

**The quaternary structure** represents the three-dimensional structure formed by aggregating two or more polypeptide chains into a single functional unit, which is then stabilized by non-covalent interactions.

In this part of the project we will interest ourselves mostly in the secondary structure and how to predict it just by using statistical information from a set of proteins with a well known one. The need for such a predictor comes from the fact that the instruments needed in order to experimentally determine it (cf. the secondary structure) are costly and also require a certain set o skills to operate them. We will further present in this project an implementation of the **GOR III method** and an analysis of this method using Q3 and MCC. This implementation will be a **three-state predictor**, meaning that it can predict three different secondary structure types: *Alpha Helices*, *Beta Strands* and *Coils*.

### The displaying functions

To make the results more readable, we will take advantage of the Jupyter Notebook's ability to display **HTML** formatted code in order to display colored text. Choosing this instead of Pythons's formatted output (__C__ like formatted output indeed) is easier and gives us the ability to use a wider range of colors. The following function, as the name suggests, displays the text in the specified color **on a new line** . This limitation (of not being able to continue writing on the same line) comes from the fact that each call to the **display** function will generate a different output block in Jupyter Notebook and will force us to define more output functions.

```python
def print_color(string, color='black') :
    """ Printing colored text using HTML. The color can be specified either by
        its name, either by an integer value (any integer is accepted). If no
        color is specified, the string will be displayed with black.
    """
    if type(color) == type(1) :
        if color < 0 or color > 11 :
            color = color % 12 #any integer will be reduced to a valid valid
        display(html('<font face="Courier" color="{}">{}</font>'.format(html_colors[color],string)))
    else :
        display(html('<font face="Courier" color="{}">{}</font>'.format(color,string)))
```

```python
def structure_color(element,graphic=False) :
    """ Generates the HTML code to print a secondary structure element or displays
        and exception and return a symbolic character #.
    """
    graphics = {True:{'H':'▥','E':'◼','C':'▬'},False:{'H':'H','E':'E','C':'C'}}
    
    if element == 'H' :
        return '<font face="Courier" color="{}">{}</font>'.format(html_colors[3],graphics[graphic]['H'])
    elif element == 'E' :
        return '<font face="Courier" color="{}">{}</font>'.format(html_colors[6],graphics[graphic]['E'])
    elif element == 'C' :
        return '<font face="Courier" color="{}">{}</font>'.format(html_colors[11],graphics[graphic]['C'])
    else :
        print_color('Error: unrecognised structure symbol {}'.format(element),'red')
        return '<font face="Courier" color="{}">{}</font>'.format('red','#')
```

The following function is used to display the secondary structure in colors. In order to improve the displaying form factor, *max_line_len* can be used to specify the maximum number of elements to be displayed on one line, the other being displayed on successive lines.

```python
def print_structure(structure, max_line_len=None, graphic=False) :
    """ Displaying a given secondary structure with colors. 
    """
    if max_line_len == None :
        max_line_len = len(structure)
    text = ''
    for i in range(0,len(structure),max_line_len) :
        #split the output in line of max_line_len length
        for pos in structure[i:i+max_line_len] :
            #determine the color for each position in the structure
            text += structure_color(pos,graphic)
        text += '<br>'
    display(html(text))
```

Similary to the previous one, the following function allows us to visualize the protein structure, but this time it will display two structures, rather than just one, in a comparative manner. This comparative view is inspired by the displaying format used for sequence alignments, and the symbol intercalated between two successive lines represents a match (':' by default, but can be changed) or a mismatch ('.' by default, but can be changed). The same as for the previous functions, the *max_line_len* attribute indicated the maximum size of the structure to be displayed on one line, while the newly introduced *preprocess* attribute modify the function behaviour so that instead of displaying the result right during the function execution, the desired output is returned as a list, each element of the list representing a distinct line of output.

```python
def print_structures(structure1, structure2, preprocess=False, max_line_len=80) :
    """ Display the structures given in arguments in a comparative manner,
        similar to the way we previously displayed the sequence alignments.
        preprocess = return list with the preprocessed result rather than
                     displaying it right away
    """
    lines = ''
    match_char = {True:':',False:'.'} #True indicates the value to be shown between
                                      #2 identical positions in the structures
    preprocessed_output = list()
    if len(structure1) == len(structure2) :
        line1 = ''
        line2 = ''
        for i in range(0,len(structure1),max_line_len) :
            #display the max part of the structure's comparative view that fits on one line
            #compute the color displaying of the structures
            for pos in structure1[i:i+max_line_len] :
                line1 += structure_color(pos)
            for pos in structure2[i:i+max_line_len] :
                line2 += structure_color(pos)
            lines += line1 + '<br>' + line2 + '<br>' 
            if preprocess :
                match = ''
                for j in range(i,min(i+max_line_len,len(structure1))) :
                    match += match_char[structure1[j] == structure2[j]]
                #create the formatted output
                preprocessed_output.append({'html':'<font face="Courier" color="grey">Predict </font>' + line1 + '<br>' \
                                           '<font face="Courier" color="white">########</font>' + \
                                           '<font face="Courier" color="grey">{}</font>'.format(match) + \
                                           '<br><font face="Courier" color="grey">Correct </font>' + line2,\
                                           'len':min(80,len(structure1)-i)})
            #pass to the next lines
            line1 = ''
            line2 = ''
        if not preprocess :
            display(html(lines))
        else :
            return preprocessed_output
    else :

        print_color('Error: Different length sequences!','red')
```

### The *DSSP* abstract data type

The **DSSP** abbreviation stands for *<b>D</b>efine <b>S</b>econdary <b>S</b>tructure of <b>P</b>roteins* and is the standard method to assing a secondary structure to a primary structure sequence of amino acids residus.

**DSSP** represents the first abstract data type to be defined for this project and is the ADT which will be used to store the information found in the multiple *.dssp* files given, in a dictionary, as previously mentioned. It is similar to the *Sequence* ADT used in the first 2 parts of this project, the difference being the fact that, besides the **residus sequence**, this data type will also store the **secondary structure** type in which the residu takes part. Its string representation respects the *fasta* format and it will be used to actually store the parsed information in the secondary file, used in order to minimize the time required to reexecute all the code. The getter of this ADT returns a couple containing the specific primary and secondary structure respectively for a given position.

```python
class DSSP :
    """ Abstract data type aggregating an amino acids residus sequence and the 
        secondary structure corresponding to this sequence.
    """
    def __init__(self, acids, structure, description = '') :
        self._description = description.rstrip('\n')
        self._acids = acids
        self._structure = structure
    def __str__(self, max_line_len=80) :
        redable = '' # Will contain the lines representing the primary and the secondary
                     # structure, intercaled, with a length of maximum max_line_len chars
                     # and respecting the fasta format
        for segment in range(0, self.len(), 80) :
            redable += self._acids[segment:segment+max_line_len] + '\n'
            redable += self._structure[segment:segment+max_line_len] + '\n'
        return self._description + '\n' + redable 
    def __getitem__(self, position) :
        return [self._acids[position],self._structure[position]]
    def len(self) :
        return len(self._acids)
    def pretty_print(self, max_line_len=80) :
        print(self._description)
        print()
        for segment in range(0, self.len(), max_line_len) :
            print(self._acids[segment:segment+max_line_len])
            print_structure(self._structure[segment:segment+max_line_len])
```

```python
def parse_dssp(file = file_name, path = dataset_path, dssp_path = dssp_path) :
    """ Parse the dssp files indicated in the reference text file
        and the chain specified in the same place.
    """
    with open (path + '/' + file, 'r') as files_list :
        for line in files_list :
            line = line.split()
            name = line[0][:-1]
            chain = line[0][-1]
            process_dssp(name, chain, path + '/' + dssp_path)
    write_to_file()
```

Given the **different structure of the various _.dssp_ files** used in this project, a parser could not be implemented using only the split function for each line in each of the files. This is the reason why we chose to use a hybrid method, combining line splitting and retrieval from a specific position. The main problem with the splitting method is that some columns are not separated by a blank space and also all the spaces are erased, including the ones which are actually a symbol for a certain type of secondary structure, namely the **Coils**, while the position retrieval method might impose some problems in terms of comprehensiveness, since the positions on the line is not always very intuitive, nor suggestive, compared to the number of the column which is analyzed. On the other hand, the hybrid method implemented here conserves some of the ease of understanting of the splitting method, while patching its weaknesses with the strength supplied by the position retrieval method. 

A few remarks regarding the files' reading and parsing process concern the need to **assimilate all the lowercase letters to Cysteine** and the need to ignore, if present on the position on which the amino acid residu is usually placed, the exclamation mark.

```python
def process_dssp(name, chain, path, extension=dssp_extension) :
    """ Process the contents of a .dssp file and save the corresponding
        DSSP type object in the dictionary.
    """
    current_chain = None
    description = None
    acids = ''
    structure = ''
    header_dump = [] #will be a list of the lines composing the heather of the dssp file
    with open(path + '/' + name + extension) as file : #automatically close the file
        for line in file :
            listed = line.split()
            if (listed[-1] != '.' and listed[0] != '#' and line[13] not in to_ignore or '!*' in line) :
                #header merker or end of chain marker
                if description == None :
                    description = '> {}{} {}'.format(name, chain, process_header(header_dump))
                if current_chain == None :
                    current_chain = line[11] #position retrieval method used to get the currently reading chain
                if '!*' in line :
                    #chain ending symbol
                    if current_chain == chain :
                        #if the chain is the desired one, save the DSSP object to dictionary
                        if name in dssp_dict :
                            dssp_dict[name][chain] = DSSP(acids, structure, description)
                        else :
                            dssp_dict[name] = {chain:DSSP(acids, structure, description)}
                    #pass to the next chain
                    current_chain = None
                    acids = ''
                    structure = ''
                else :
                    #all the amino acids in lowercase letters are assimilated to Cysteine
                    if line[13] >= 'a' and line[13] <= 'z' :
                        acids += 'C'
                    else :
                        acids += line[13]
                    structure += sec_structure[line[16]]
            else :
                header_dump.append(listed)
    if current_chain == chain :
        #save the last read chain, since it does not end with the typical ending symbol
        if name in dssp_dict :
            dssp_dict[name][chain] = DSSP(acids, structure, description)
        else :
            dssp_dict[name] = {chain:DSSP(acids, structure, description)}
```

Parsing the DSSP information from the 3713 given files might take a while (usually the processing time is around one minute) because of the high percentage of unused information in this files compared to the actually used information. From a line of 136 characters, we usually only use the three representing the **chain**, the **amino acid residu** and the **secondary struture**. This leads to about 2% of the processed information to be actually used, while the other is just a loss of processing time. In order to avoid this loss, we decided to store the result of the parsing, in the fasta format, in a new file. This way, while running the notebook again we obtain a 100% use of the parsed information (no character is discarded, besided line return, but it might be ignored, since it will not be present if we set the maximum length of the information to be written on a line to infinity) and a dramatically shorter parsing time, which is basically less than a second.

In order to save the *DSSP* objects to a file using the fasta format, we will append to the desired file the string representation of that object, which is returned by the previously defined method *\_\_str\_\_*.

```python
def write_to_file(file_name = fasta_format_output) :
    """ Saves the contents of the dictionary to a chosen output file
        in the fasta format. Should be called after the parsing process
        took place and the dictionary was populated.
    """
    open(file_name, 'w').close() # erase the content of the file before writing anything
    with open(file_name, 'a') as fasta_file :
        for protein in dssp_dict :
            for chain in dssp_dict[protein] :
                fasta_file.write(str(dssp_dict[protein][chain])+'\n')
```

As already specified, to speed up the following parsing processes, a file will be created, containing the result of the original parsing of the 3713 _.dssp_ files. This file can then be used to build the working dictionary (*dssp_dict*), since parsing from the custom fasta file takes significantly less time than the original parsing, because of the reduction in the percentage of unused, but processed, data. This second type of parsing is similar to the one used in the previous parts, reading the fasta file line by line and creating the *DSSP* object when the reading was complete, or in other words, either when the header of a new structure is encountered, either when the entire file was read. To continue this analogy with the previously implemented code, the only difference is that the residus sequence lines are intercalated with the corresponding secondary structure lines in the text file.

In this project, in order to obtain a good performance and to also prove the working state of both parsers, we will use the fasta parser for the bigger dataset, while the classic parser is used for the smaller, the test dataset. We should still mention that the fasta file was built initially using the previously presented parser.

```python
def parse_dssp_fasta(file_name=fasta_format_output) :
    """ Fasta parsing used instead of the regular parsing to reduce the execution time.
        
    """
    with open(file_name, 'r') as fasta_file :
        name = None
        for line in fasta_file :
            if line[0] == '>' :
                #header of a new structure
                if name != None :
                    #not the first structure, so there is something to be built
                    if name[:-1] not in dssp_dict :
                        dssp_dict[name[:-1]] = dict()
                    dssp_dict[name[:-1]][name[-1]] = DSSP(acids, structure, description)
                #clear the variables in order to build the new structure
                name = line.split()[1]
                description = line.strip()
                acids = ''
                structure = ''
                is_sequence_line = True # True if next line will contain a part of the primary structure
                                        # and False if it will contain part of the secondary structure
            else :
                #reading a line if the body of the structure (either an amino acids sequence,
                #either a secondary structure)
                if is_sequence_line :
                    is_sequence_line = not is_sequence_line
                    acids += line.strip()
                else :
                    is_sequence_line = not is_sequence_line
                    structure += line.strip()
        #save the last structure in the dictionary
        if name[:-1] not in dssp_dict :
            dssp_dict[name[:-1]] = dict()
        dssp_dict[name[:-1]][name[-1]] = DSSP(acids, structure, description)
```

```python
def process_header(header_dump) :
    """ Compute the second part of the fasta header of a protein
        (containing the protein name and the source organism) from
        the .dssp file header.
    """
    protein = 'unknown'
    source = 'unknown'
    for line in header_dump :
        if line[0] == 'HEADER' :
            protein = ' '.join(line[1:-3])
        elif line[0] == 'SOURCE' :
            source = ' '.join(line[3:-1])[:-1]
    return '| {} | {}'.format(protein, source)
```

```python
def compute_dataset(include_chain=False, include_type=False, file=file_name, path=dataset_path) :
    """ Returns a list containing the proteins' names in the input file 
        which can be either 'CATH_info.txt', either 'CATH_info_test.txt'
        or any other file with the same structure.
    """
    proteins = list()
    with open(path + '/' + file,'r') as protein_list :
        for line in protein_list :
            line = line.split()
            if include_chain :
                proteins.append(line[0])
            else :
                proteins.append(line[0][:-1])
            if include_type :
                proteins[-1] += line[1]
    return proteins
```

### The GOR III method

*GOR* is an acronym for the names of the researchers who developed this method: *<b>G</b>arnier, <b>O</b>sguthorpe and <b>R</b>obson*. The **GOR method** is an information based theory for secondary structure prediction in proteins and over the years multiple versions of this method were developed, improving its accuracy. In this paper we will implement *GOR III*, the third iteration of the original method, which has improved accuracy, but its fundaments are still based on the **Bayesian statistics** and the **information theory** developed in the middle of the XXth century. The main differences between the first version of GOR and the one we are going to implement are the number of struture types predicted, since the first implementation used to predict four different structures, *H*, *E*, *C* and *T*, while the last one eliminated one of them, the *T* (turns), which is not predicted anymore, the fact that **GOR III takes into account the correlation between the type of residus in the window and the type of residu to be predicted**, while GOR I did not, and, of course, the database used.

```python
class GOR3() :
    """ Abstract data type used to implement the GOR III method containing the
        data variables which store the information found through training and
        the methods used to train the prediction and make a new one.
    """
    def __init__(self) :
        amino_acids = 'ACDEFGHIKLMNPQRSTVWY'
        self.negation = {'C':['H','E'],'H':['C','E'],'E':['C','H']}
        self.freq = dict()
        for i in self.negation :
            self.freq[i] = dict()
            self.freq[i][0] = 0
            for j in amino_acids :
                self.freq[i][j] = dict()
                self.freq[i][j][0] = 0
                for k in amino_acids :
                    self.freq[i][j][k] = 0
```

```python
trained_gor = GOR3()
```

```python
def count_residus(files_list=compute_dataset(), files_number=3000) :
    """ Return the number of residus in the trainig set.
    """
    already_processed = list()
    count = 0
    for protein in files_list :
        if len(already_processed) == files_number :
            #stop after the specified number of proteins is reached
            break
        if protein not in already_processed :
            already_processed.append(protein)
            for chain in dssp_dict[protein] :
                #process all the chains (to be used) of the protein
                count += dssp_dict[protein][chain].len()
    return count

print('The training set contains {} residus.'.format(count_residus()))
```

As we can see in the code below, the **training** process for the GOR method represents just the creation of the diffrent required frequencies by consulting the training dataset. Due to the limitations in the dataset availale at that time, which contained about 63000 residus, because some rare amino acids (like *Tryptophan* or *Methionine*) had a really small frequency, **data augumentation tachniques** were used, but this will not be the case for our project, since the number of amino acids in out training set (642579) is roughly ten times bigger than the original one.

```python
def count_sequence(self, sequence) :
    """ Train the method to take into consideration a new sequence with its
        secondary structure.
    """
    for pos in range(sequence.len()) :
        aa, ss = sequence[pos]
        self.freq[ss][0] += 1     #increment the number of times we found the
                                  #that secondary structure (ss)      
        self.freq[ss][aa][0] += 1 #increment the number of times we found the
                                  #the structure ss for the amino acid aa       
        #create a list with the neighbours of the analysed position
        neighbourhood = [neighbour for neighbour in range(max(pos-8,0),min(pos+9,sequence.len()))]
        neighbourhood.remove(pos)
        
        for neighbour in neighbourhood :
            #increment the number of times we found the structre ss for the
            #amino acid aa when the specific neighbour is found
            self.freq[ss][sequence[neighbour][0]][aa] += 1
            
GOR3.count_sequence = count_sequence

def feed_sequences(self, files_list=compute_dataset(), files_number=3000) :
    """ Feed a specified number (files_number) of sequences to the training
        method of GOR. The sequences are taken from a list given as paramater.
    """
    already_processed = list()
    for protein in files_list :
        if len(already_processed) == files_number :
            #stop after the specified number of proteins is feeded
            break
        if protein not in already_processed :
            already_processed.append(protein)
            for chain in dssp_dict[protein] :
                #process all the chains (to be used) of the protein
                self.count_sequence(dssp_dict[protein][chain])
                
GOR3.feed_sequences = feed_sequences
GOR3.train = feed_sequences
```

The individual information represents the likelyhood of an amino acid (R) residu to be part of a certain secondary structure (S) and can be calculated using the following formula :

$$ I(\varDelta S;R) = I(S; R) - I(nonS;R)\\
   I(\varDelta S;R)=log(\frac{f_{S;R}}{f_{nonS;R}}) + log(\frac{f_{nonS}}{f_S}) $$
   
This formula uses *non S* structures to increase the information on *S* structures and can be seen as a kind of normalization, since the number of amino acids and the number of residus is not taken into consideration.

We use here the word likelyhood, rather than probability, because it fits better the concept described in this section, since the result of the *log* function used in the formula above might be negative, hence violating the probability property that it is a positive number smaller than unity.

```python
def individual_info(self,ss,aa) :
    """ The likeliness of an amino acid (aa) to be part of a certain 
        structure (ss).
    """
    return log(self.freq[ss][aa][0]/\
               (self.freq[self.negation[ss][0]][aa][0] + self.freq[self.negation[ss][1]][aa][0])) +\
           log((self.freq[self.negation[ss][0]][0] + self.freq[self.negation[ss][1]][0])/\
               self.freq[ss][0])
    
GOR3.individual_info = individual_info
```

The pairing (couple) information represents the likelyhood of an amino acid residu to be part of a certain secondary structure, taking into account what its neighbour is, and it is calculated using the following formula :
$$ I(\varDelta S_j;R_{j+m}|R_j) = log(\frac{f_{S_j,R_{j+m},R_j}}{f_{nonS_j,R_{j+m},R_j}}) 
                                + log(\frac{f_{nonS_j,R_j}}{f_{S_j,R_j}}) $$

```python
def couple_info(self,ss,r1,r2) :
    """ The likeliness of an amino acid (r2) to be part of a certain 
        structure (ss) taking into account his neighbour (r1).
    """
    return log(self.freq[ss][r1][r2]/\
              (self.freq[self.negation[ss][0]][r1][r2] + self.freq[self.negation[ss][1]][r1][r2])) +\
           log((self.freq[self.negation[ss][0]][r2][0] + self.freq[self.negation[ss][1]][r2][0])/\
               self.freq[ss][r2][0])
    
GOR3.couple_info = couple_info
```

The neighbourhood information can be described as a sum of multiple couple informations, as revealed by its computation formula :

$$ \sum_{m=-8 \\ m \neq 0}^{m=8} I(\varDelta S_j;R_{j+m}|R_j) $$

The window chosen is eight positions long at the left and at the right of the in-treatment position. If less than eight positions exist on either side, we will take into account the maximum number of positions available, hence the size of the lateral windows varying between zero (at the sequence extremities) and eight neighbours. The number eight was chosen based on studies of information content at increasing separations.

```python
def neighbourhood_info(self, sequence, position, ss) :
    """ The likeliness of an amino acid (aa) to be part of a certain 
        structure (ss) taking into account his neighbours, eight of
        them at the left and eight at the right (if existing).
    """
    aa = sequence[position][0]
    total = 0
    neighbourhood = [j for j in range(max(position-8,0),min(position+9,sequence.len()))]
    neighbourhood.remove(position)
    for j in neighbourhood :
        neighbour = sequence[j][0]
        total += self.couple_info(ss,neighbour,aa)
    return total

GOR3.neighbourhood_info = neighbourhood_info
```

In order to predict the secondary structure of a new sequence, the GOR III uses the statistics created by its training. For each position, the predicted structure is the one maximizing the following equation :

$$ I(\varDelta S_j;R_1 ... R_n) \approx I(\varDelta S_j;R_j) + \sum_{m \neq 0} I(\varDelta S_j;R_{j+m}|R_j) $$

```python
def make_prediction(self,sequence) :
    """ Returns a dictionary containing the predicted secondary structure of
        the sequence given in arguments and the scores for each position and
        for each of the three possible structure types.
    """
    score = {'H':list(),'E':list(),'C':list()}
    res = list()
    for position in range(sequence.len()) :
        aa = sequence[position][0]
        for ss in score :
            score[ss].append(self.individual_info(ss,aa) + self.neighbourhood_info(sequence,position,ss))
        if score['H'][position] >= score['E'][position] and score['H'][position] >= score['C'][position] :
            res.append('H')
        elif score['E'][position] >= score['H'][position] and score['E'][position] >= score['C'][position] :
            res.append('E')
        else :
            res.append('C')
    score['res'] = res
    return score

GOR3.make_prediction = make_prediction
```

### GOR III evaluation

In order to evaluate our prediction we will use two different accuracy measurements: **Q3** and **MCC**. The first of them, Q3, is calculated in a similar manner to the way a probability is calculated :

$$ Q_3 = \frac{N_{residus\ correctly\ predicted}}{N_{residus\ in\ total}} $$

For both Q3 and MCC we will calculate the average and the deviation from this average using the following formulas :

$$ \mu=  \frac{1}{n}\sum_{1}^{p}n_i*x_i $$
$$ s = \sqrt{ \frac{1}{n}\sum_{1}^{p}n_i(x_i - \mu)^2  } $$

```python
def q3(expected, predicted) :
    """ Compute the Q3 measure for one prediction.
    """
    correct = 0
    for i in range(len(expected)) :
        if expected[i] == predicted[i] :
            correct += 1
    return correct/len(expected)
```

```python
def q3_average_deviation(file=file_name, path=dataset_path,a=trained_gor) :
    """ Compute the Q3 average and deviation using the 713
        test proteins.
    """
    avgq3 = 0
    count = 0
    results = list()
    deviation = 0
    dataset = compute_dataset(file=file_name, path=dataset_path)
    already_used = list()
    for i in range(3000, len(dataset)) :
        if dataset[i] not in already_used :
            already_used.append(dataset[i])
            for chain in dssp_dict[dataset[i]] :
                results.append(a.make_prediction(dssp_dict[dataset[i]][chain]))
                count += 1
                results[-1]['q3'] = q3(results[-1]['res'],dssp_dict[dataset[i]][chain]._structure)
                avgq3 += results[-1]['q3']
    avgq3 = avgq3/count
    for result in results :
        deviation += (result['q3'] - avgq3)**2
    deviation = deviation/count
    deviation = sqrt(deviation)
    return avgq3, deviation
```

**MCC** is the acronym for **Matthews correlation coefficient**, which is a measurement for the quality of binary predictions and is considered to be one of the best ways to describe a **confusion matrix**. The MCC can be calculated using the following formula :

$$ MCC = \frac{TP*TN - FP*FN}{\sqrt{(TP+FP)(TP+FN)(TN+FP)(TN+FN)}} $$

The first letter in each couple represents the positivity of the **condition**, while the second letter represents the truth of the **prediction**.

```python
def mcc(ss, expected, predicted, testing=False) :
    """ Compute the MCC measure for one prediction.
    """
    negation = {'H':'EC','E':'HC','C':'EH'}
    conf_mat = [[0,0],[0,0]] # TP = conf_mat[1][1] = expected ss and predicted ss
                             # TN = conf_mat[1][0] = expected ~ss and predicted ~ss
                             # FP = conf_mat[0][1] = expected ss and predicted ~ss
                             # FN = conf_mat[0][0] = expected ~ss and predicted ss
    #build the confusion matrix
    for i in range(len(predicted)) :
        prediction = int(predicted[i] == ss)
        if prediction :
            condition = int(predicted[i] == expected[i])
        else :
            condition = int(expected[i] in negation[ss])
        conf_mat[condition][prediction] += 1
        
    #solve division by zero problem
    if (conf_mat[1][1]*conf_mat[1][0] - conf_mat[0][1]*conf_mat[0][0]) == 0 :
        if not testing : 
            return 0
        else :
            return conf_mat[1][1],conf_mat[0][1],conf_mat[1][0],conf_mat[0][0]
    else :
        if conf_mat[1][1]+conf_mat[0][1] == 0 :
            conf_mat[1][1] = 1
        if conf_mat[1][1]+conf_mat[0][0] == 0 :
            conf_mat[1][1] = 1
        if conf_mat[1][0]+conf_mat[0][1] == 0 :
            conf_mat[1][0] = 1
        if conf_mat[1][0]+conf_mat[0][0] == 0 :
            conf_mat[1][0] = 1
        if testing :
            return conf_mat[1][1],conf_mat[0][1],conf_mat[1][0],conf_mat[0][0]
        else :
            return (conf_mat[1][1]*conf_mat[1][0] - conf_mat[0][1]*conf_mat[0][0])/\
                   sqrt((conf_mat[1][1]+conf_mat[0][1])*(conf_mat[1][1]+conf_mat[0][0])*\
                        (conf_mat[1][0]+conf_mat[0][1])*(conf_mat[1][0]+conf_mat[0][0]))
```

```python
def mcc_average_deviation(file=file_name, path=dataset_path, a=trained_gor) :
    """ Compute the MCC average and deviation for the 713
        test proteines.
    """
    avgmcce = 0; avgmcch = 0; avgmccc = 0
    count = 0
    results = list()
    deviation_mcce = 0; deviation_mccc = 0; deviation_mcch = 0
    dataset = compute_dataset(file=file_name, path=dataset_path)
    already_used = list()
    for i in range(3000, len(dataset)) :
        if dataset[i] not in already_used :
            already_used.append(dataset[i])
            for chain in dssp_dict[dataset[i]] :
                results.append(a.make_prediction(dssp_dict[dataset[i]][chain]))
                count += 1
                results[-1]['mcce'] = mcc('E',results[-1]['res'],dssp_dict[dataset[i]][chain]._structure)
                results[-1]['mcch'] = mcc('H',results[-1]['res'],dssp_dict[dataset[i]][chain]._structure)
                results[-1]['mccc'] = mcc('C',results[-1]['res'],dssp_dict[dataset[i]][chain]._structure)
                avgmcce += results[-1]['mcce']; avgmccc += results[-1]['mccc']; avgmcch += results[-1]['mcch']
    avgmcce = avgmcce/count; avgmccc = avgmccc/count; avgmcch = avgmcch/count
    for result in results :
        deviation_mcce += (result['mcce'] - avgmcce)**2
        deviation_mccc += (result['mccc'] - avgmccc)**2
        deviation_mcch += (result['mcch'] - avgmcch)**2
    deviation_mcce = deviation_mcce/count; deviation_mccc = deviation_mccc/count; deviation_mcch = deviation_mcch/count
    deviation_mcce = sqrt(deviation_mcce); deviation_mccc = sqrt(deviation_mccc); deviation_mcch = sqrt(deviation_mcch)
    return avgmcce, avgmccc, avgmcch, deviation_mcce, deviation_mccc, deviation_mcch
```

## Proof of concepts and results

Before doing a more complex analysis of our GOR III method implementation, we will run some tests to assure the correctness of the parsed data and of the confusion matrix computation. To test the correctness of the parsed data, we will make sure that every amino acid sequence saved in the dictionary contains on each position one of the 20 symbols for genetically encoded proteinogenic amino acids. This test was extremely usefull while debugging the parser and helped discover some of the problems the *.dssp* impose, for example, the '!' which is placed in the column of residus in one of files. For the second test, we will compare the results of our method to build the confusion matrix with the results computed by hand. Note that some of the results might be different that the ones presented on the slide due to the fact that, when computing the MCC, we chose to use a different method, more explicitly, when the numerator in the MCC formula is equal to zero, instead of calculating the denominator, we are returning directly zero. This variation of the method is superior because computing the denominator might lead to the zero value, case in which the values of the confusion matrix will be alterated to avoid the division by zero. This alteration would lead to big differences in some cases. For example, computing the MCC for the fifth protein in the testing test with the original method for a Beta Strand structure would give a score of about 0.13, while the second method gives, obviously, the score 0, which is more adequate.

```python
def test_parsing(file=file_name, path=dataset_path) :
    """ Test the parsing.
    """
    error_p = False
    with open(path + '/' + file,'r') as protein_list :
        for line in protein_list :
            line = line.split()
            if line[0][:-1] not in dssp_dict or line[0][-1] not in dssp_dict[line[0][:-1]] :
                error_p = 'Error protein {} - chain {} : not parsed!'.format(line[0][:-1], line[0][-1])

    error_aa = False
    error_ss = False
    aa = 'ACDEFGHIKLMNPQRSTVWY'; ss = 'CEH'
    for i in dssp_dict :
        for j in dssp_dict[i] :
            for h in range(dssp_dict[i][j].len()) :
                if dssp_dict[i][j][h][0] not in aa :
                    error_aa = 'Error protein {} - chain {} : wrong amino acid'.format(i, j)
                if dssp_dict[i][j][h][1] not in ss :
                    error_ss = 'Error protein {} - chain {} : wrong structure'.format(i, j)
    
    if error_p : print(error_p)
    if error_aa : print(error_aa)
    if error_ss : print(error_ss)
    if not error_p and not error_aa and not error_ss : print('Test successful: no parsing error found.')
        
test_parsing()
```

```python
def test_mcc() :
    """ Test the if the confusion matrix is correctly built.
    """
    error = False
    temp = mcc('H','HHHHHCCCCEEEECCCEEECCCHHHH','CHHHCCCCEEEECCCEEECCCHHHHC',testing=True)
    if temp[0] != 6 or temp[1] != 1 or temp[2] != 16 or temp[3] != 3 :
        print_color('Error: confusion matrix not correct!','red')
        error = True
    temp = mcc('E','HHHHHCCCCEEEECCCEEECCCHHHH','CHHHCCCCEEEECCCEEECCCHHHHC',testing=True)
    if temp[0] != 5 or temp[1] != 2 or temp[2] != 17 or temp[3] != 2 :
        print_color('Error: confusion matrix not correct!','red')
        error = True
    temp = mcc('H','HHHHHCCCCEEEECCCEEECCCHHHH','HHHHCCCCCHHHHCCCHHCCCCHHHH',testing=True)
    if temp[0] != 8 or temp[1] != 6 or temp[2] != 11 or temp[3] != 1 :
        print_color('Error: confusion matrix not correct!','red')
        error = True
    temp = mcc('E','HHHHHCCCCEEEECCCEEECCCHHHH','HHHHCCCCCHHHHCCCHHCCCCHHHH',testing=True)
    if temp[0] != 0 or temp[1] != 0 or temp[2] != 19 or temp[3] != 7 :
        print_color('Error: confusion matrix not correct!','red')
        error = True
    if not error :
        print('Test successful: confusion matrix correctly built.')
        
test_mcc()
```

```python
def check_multiple_chains(file=file_name, path=dataset_path) :
    """ Displays the proteins from which we are going to use
        multiple chains. Used for testing.
    """
    prot = dict()
    duplicate = dict()
    with open(path + '/' + file,'r') as protein_list :
        for line in protein_list :
            line = line.split()
            if line[0][:-1] in prot :
                if line[0][:-1] in duplicate :
                    duplicate[line[0][:-1]].append(line[0][-1])
                else :
                    duplicate[line[0][:-1]] = list(line[0][-1])
            else :
                prot[line[0][:-1]] = line[0][-1]
    if len(duplicate) == 0 :
        print('Only using one chain for each of the proteins in the dataset.')
    else :
        print('Using multiple chains of the following proteins: ')
        for protein in duplicate.keys() :
            duplicate[protein].insert(0,prot[protein])
            print('- protein {} - chains : {}'.format(protein,', '.join(duplicate[protein])))
```
